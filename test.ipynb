```python
from datetime import datetime, time


# 方法1: 使用datetime计算
def get_nanoseconds_from_midnight(target_time: datetime) -> int:
    """计算从今天0点到指定时间的纳秒数"""
    # 获取今天的0点
    midnight = datetime.combine(target_time.date(), time.min)

    # 计算时间差
    time_diff = target_time - midnight

    # 转换为纳秒 (1秒 = 1,000,000,000纳秒)
    return int(time_diff.total_seconds() * 1_000_000_000)


# 方法2: 直接使用时间组件计算
def get_nanoseconds_from_time_components(
    hour: int, minute: int, second: int, microsecond: int = 0
) -> int:
    """根据时分秒微秒计算从0点开始的纳秒数"""
    total_seconds = hour * 3600 + minute * 60 + second
    total_microseconds = total_seconds * 1_000_000 + microsecond
    return total_microseconds * 1000  # 转换为纳秒


# 示例使用
now = datetime.now()
nanoseconds_from_midnight = get_nanoseconds_from_midnight(now)
print(f"从今天0点到现在的纳秒数: {nanoseconds_from_midnight}")

# 示例: 10:00:00的纳秒数
nanoseconds_10am = get_nanoseconds_from_time_components(10, 0, 0)
print(f"从0点到10:00:00的纳秒数: {nanoseconds_10am}")

# 验证结果
print(f"10小时 = {10 * 3600 * 1_000_000_000} 纳秒")
```
